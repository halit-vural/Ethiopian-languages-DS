```python
import requests
import json
from bs4 import BeautifulSoup

url = 'https://www.bbc.com/amharic'

# Use requests to fetch the webpage content
response = requests.get(url)

# Use BeautifulSoup to parse the webpage HTML
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the links to articles on the webpage
article_links = []
for link in soup.find_all('a'):
    href = link.get('href')
    if href and href.startswith('/amharic/articles'):
        article_links.append('https://www.bbc.com' + href)

# Loop through the article links and extract the information for each article
articles = []
for article_link in article_links:
    # Use requests to fetch the article content
    article_response = requests.get(article_link)

    # Use BeautifulSoup to parse the article HTML
    article_soup = BeautifulSoup(article_response.content, 'html.parser')

    # Extract the headline, date, views, categories, and content of the article
    article_title = article_soup.find('h1').text.strip()
    article_date = article_soup.find('time').text.strip()
    #article_views = article_soup.find('span', class_='qa-story-total-views').text.strip()
    article_categories = [cat.text.strip() for cat in article_soup.find_all('a', class_='qa-story-theme-link')]
    article_url = article_link

    # Create a dictionary for the article and append it to the list of articles
    article_dict = {
        'title': article_title,
        'date': article_date,
        #'views': article_views,
        'categories': article_categories,
        'url': article_url
    }
    articles.append(article_dict)

# Write the list of articles to a JSON file
with open('articles.json', 'w') as f:
    json.dump(articles, f)



# Load the data from the JSON file
with open('articles.json', 'r') as f:
    articles = json.load(f)

# Loop through the articles and extract the headline and date
for article in articles:
    headline = article['title']
    date = article['date']
    
    # Do something with the headline and date
    print(headline, date)
```
